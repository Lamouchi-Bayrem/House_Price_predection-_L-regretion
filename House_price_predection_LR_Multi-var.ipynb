**Machine learning with python : linear regression with multiple variables**

House price depends on area (square feet), bed rooms and age of the home (in years). Given these prices we have to predict prices of new homes based on area, bed rooms and age.

1. Importing Libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('/homeprices.csv')
df
```

3. Data Cleaning:

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

4. Data Visualization

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['area'], df['price'])
plt.xlabel('Area (sqft)')
plt.ylabel('Price ($)')
_ = plt.title('Price vs Area')
```

5. Linear Regression:

```python
price = df.price
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('price',axis='columns'),df.price)
```

6. Manual Price Calculation:

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2500,2,52]])
```
